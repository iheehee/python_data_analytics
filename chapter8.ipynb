```python
import pandas as pd
import seaborn as sns

mpg = pd.read_csv('data/mpg.csv')
sns.scatterplot(data = mpg, x = 'cty', y= 'hwy')
```

```python
midwest = pd.read_csv('data/midwest.csv')
sns.scatterplot(data = midwest, x = 'poptotal', y= 'popasian').set(xlim = (0, 500000), ylim = (10, 10000))
```

```python
df = mpg.query('category == "suv"').groupby('manufacturer', as_index = False).agg(mean_cty = ('cty', 'mean')).sort_values('mean_cty', ascending = False).head()
df
```

```python
sns.barplot(data = df, x= 'manufacturer', y = 'mean_cty')
```

```python
# Q2 
df_mpg = mpg.groupby('category', as_index = False).agg(count = ('category', 'count')).sort_values('count',ascending = False)
df_mpg
```

```python
sns.barplot(data = df_mpg, x = 'category', y = 'count')
```

```python
# p.217  
# Q/1 
economics = pd.read_csv('data/economics.csv')
economics['date2'] = pd.to_datetime(economics['date'])
economics['year'] = economics['date2'].dt.year
sns.lineplot(data = economics, x='year', y = 'psavert', ci = None)
```

```python
economics['month'] = economics['date2'].dt.month
df_2014 = economics.query('year == 2014')
sns.lineplot(data=df_2014, x = 'month', y = 'psavert', ci = None)
```
